### 셀의 삽입
- 상단의 `+Code` `+Text` 버튼을 클릭
- 셀의 위/아래 `+Code` `+Text` 버튼을 클릭
- 단축키를 이용
 - 현재 셀 위에 새로운 셀 삽입 (`Ctrl + M + A`)
 - 현재 셀 아래에 새루운 셀 삽입 (`Ctrl + M + B`)

### 셀의 종류 바꾸기
- 단축키를 이용
 - 코드 셀로 변경 (`Ctrl + M + Y`)
 - 텍스트 셀로 변경 (`Ctrl + M + M`)

### 셀의 삭제
- 마우스 우클릭 > `Delete Cell`
- 삭제할 셀의 우측 상단의 휴지통 아이콘 클릭
- 단축키를 이용
 - `Ctrl + M + D`

### 셀의 실행
- 단축키를 이용
 - 현재 셀을 실행하고 현재 셀에 커서 유지 (`Ctrl + Enter`)
 - 현재 셀을 실행하고 다음 셀로 커서 옮김 (`Shift + Enter`)
