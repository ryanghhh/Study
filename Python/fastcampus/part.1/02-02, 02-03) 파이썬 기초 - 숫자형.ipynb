## **[1] 숫자형이란?**
- 숫자 형태로 이루어진 자료형

|항목|Python 표현 예시|
|:---------|:---|
|정수형 (int)|123, -123, 0|
|실수형 (float)|100.5, -100.5, 10e3, 10e5|

- 정수형

```python
a = 123
print(a)
print(type(a))
```

```python
print(type(6.3))
```

```python
b = -123
print(b)
print(type(b))
```

- 실수형

```python
a = 1.2
print(a)
print(type(a))
```

```python
b = 10e3
print(b)
print(type(b))
```

## **[4] 할당연산**

- 변수 a는 10의 값을 가질 때, 변수 a에 1을 더해서 a에 덮어씌우려면?

```python
a = 10
print(a)
a = a + 1
print(a)
```

```python
a = 10
print(a)
a += 1
print(a)
```

- 다른 사칙연산도 마찬가지로 계산할 수 있습니다.

```python
a = 10
print(a)
a += 1
print(a)
a -= 1
print(a)
a *= 2
print(a)
a /= 2
print(a)
```

```python
r = 5
print(r)
r += 1
print(r)
r -= 1
print(r)
r *= 2
print(r)
r /= 5
print(r)
```
