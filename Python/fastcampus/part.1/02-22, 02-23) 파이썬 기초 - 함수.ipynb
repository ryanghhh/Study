## **[2] 변수의 범위**
- 함수 안에서의 변수는 그 함수 안에서만 작동합니다. 함수 밖의 변수 이름과는 전혀 상관이 없습니다.

```python
x = 1

def twice(x): ## 함수안의 x는 함수 안에 수행문에서 쓰일 변수의 이름일 뿐
    x = x * 2
    return x

print(twice(x))
print(x)
```

```python
x = 1

def twice(i): ## 함수안의 i는 함수 안에 수행문에서 쓰일 변수의 이름일 뿐
    i = i * 2
    return i

print(twice(x))
print(x)
```

```python
print(i)
```

### 1. 리스트를 입력받으면 리스트 안에 있는 요소들의 평균과 분산를 반환하는 함수를 작성하여 다음 리스트의 평균과 분산을 구해보세요.

```python
ls = [5,10,29,20,40,26,15]
```

```python
average = sum(ls) / len(ls)
def average (x):
    return sum(x) / len(x)
```

```python
average(ls)
```

```python
## 선생님 풀이
## 분산
average = sum(ls) / len(ls)
vsum = sum([(i - average)**2 for i in ls])
var = vsum / len(ls)
print(var)
```

```python
def  get_average_var(x):
    average = sum(x) / len(x)

    vsum = sum([(i - average)**2 for i in x])
    var = vsum / len(x)
    
    return average, var

average, var = get_average_var(ls)
print(average)
print(var)
```

### 2. 지난 강의의 가위바위보 게임을 함수로 구현해보세요.

```python
import random
computer = random.choice(['가위', '바위', '보'])

win_dict = {'가위':'보', '바위':'가위', '보':'바위'}

player = input('가위 바위 보 중 하나를 입력하세요. ')

print(f'플레이어는 {player}를 냈습니다.')
print(f'컴퓨터는 {computer}를 냈습니다.')

if player not in win_dict:
    print('잘못된 값을 입력했습니다.')
elif computer == player:
    print('무승부')
elif win_dict[computer] == player:
    print('컴퓨터 승리')
else:
    print('플레이어 승리')
```

```python
## 선생님 풀이

import random

def rps_game(player):
    computer = random.choice(['가위', '바위', '보'])

    win_dict = {'가위':'보', '바위':'가위', '보':'바위'}

    player = input('가위 바위 보 중 하나를 입력하세요. ')

    print(f'플레이어는 {player}를 냈습니다.')
    print(f'컴퓨터는 {computer}를 냈습니다.')

    if player not in win_dict:
        print('잘못된 값을 입력했습니다.')
    elif computer == player:
        print('무승부')
    elif win_dict[computer] == player:
        print('컴퓨터 승리')
    else:
        print('플레이어 승리')

```

```python
rps_game(player)
```

### 3. 두 숫자와 연산자 기호를 입력 받으면 사칙연산을 수행하는 함수를 만들어보세요.

    예시:
    cal(34, 60, '+')

    결괏값:
    94

```python
def f(x):
    return a + c +b
```

```python
## 선생님 풀이
def cal(x, y, oper):
    if oper == '+':
        return x + y
    elif oper == '-':
        return x - y
    elif oper == '*':
        return x * y
    elif oper == '/':
        return x / y
    else:
        print('잘못된 연산자입니다.')
        
```

```python
cal(34,60, '+')
```

### 4. 다음 딕셔너리는 같이 학생들의 시험 점수입니다. 시험 점수 내림차순으로 정렬하여 학생의 이름과 점수, 그리고 등수를 출력해보세요.
- 힌트: `sorted()` 함수와 그의 인자로 `key = lambda` 함수를 이용해보세요.

```python
score = {'민준':80, '서현':75, '도아':40, '기찬':90, '재영':60}
```

```python
sorted_score = sorted(score.items(), key=lambda x: x[1], reverse = True )
```

```python
print(sorted_score)
```

```python
rank = 1
for k , v in sorted_score:
    print('이름:',k)
    print('점수:',v)
    print('순위:',rank)
    print()

    rank += 1
```
