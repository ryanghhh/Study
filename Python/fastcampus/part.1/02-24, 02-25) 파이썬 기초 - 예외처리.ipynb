## **[1] 예외처리란?**
- 오류를 처리하는 방법입니다.


    try: #try문을 실행하고
        ...
    except 발생오류 as 오류변수: #만약 에러가 나면 수행문을 실행하라.
        ...

- 10 / 0을 하면 파이썬에서 어떻게 처리할까요?
 - ZeroDivisionError가 납니다.

```python
10 / 0
```

- `try except` 문을 써서 오류 시에 해당 오류 내용을 출력하도록 해봅시다.

```python
try:
    a = 10 / 0
except ZeroDivisionError as e:
    print(e)
```

- `ZeroDivisionError`가 날 때 0이 되도록 해봅시다.

```python
try:
    a = 10 / 0
except ZeroDivisionError:
    a = 0
```

```python
a
```

- `ZeroDivisionError`가 날 때 오류 없이 코드를 넘어가도록 해봅시다.

```python
try:
    b = 10 / 0
except ZeroDivisionError:
    pass
```

```python
b
```

- 발생오류를 명시하지 않으면, 모든 오류에 대해 `except`문을 처리합니다.

```python
a = [1,2,3,4,5]
b = 6

try:
    print(a.index(b))
except:
    print('오류')
```

- 오류명을 모를 때 `Exception`으로 오류명을 출력할 수 있습니다.

```python
try:
    print(a.index(b))
except Exception as e:
    print(e)
```

```python
a = [1,2,3,4,5]
b = 6
a.index(b)
```

## **[2] `else`와 `finally`**

    try:
        ...
    except 발생오류 as 오류변수:
        ...
    else: #오류가 없을 때만 수행됩니다.
        ...
&nbsp;

    try:
        ...
    finally: #오류가 나더라도 수행됩니다.
        ...

```python
try:
    a = int(input('정수를 입력해주세요. '))
except Exception as e:
    print(e)
else:
    print(int(a) + 1)
```

```python
try:
    a = int(input('정수를 입력해주세요. '))
except Exception as e:
    print(e)
else:
    print(int(a) + 1)
```

```python
try:
    a = input('정수를 입력해주세요. ')
except Exception as e:
    print(e)
else:
    print(int(a) + 1)
```

```python
try:
    a = 1
    b = a/0
finally:
    print('종료')
```

```python
try:
    a = input('정수를 입력해주세요. ')
except Exception as e:
    print(e)
else:
    print(int(a) + 1)
finally:
    print('종료')
```

### 1. 숫자 맞추기 게임을 함수로 만들어보세요.
- 맞출 숫자는 아래 코드와 같이 1~100까지 자연수 중 랜덤으로 정합니다.
- 숫자를 입력할 때마다 더 큰지 작은지 여부를 말해줍니다.
- 1~100까지의 정수가 아닌 정수가 입력된다면 '1부터 100까지의 정수 중 하나를 입력하세요.'라고 출력하고 게임을 계속 이어갑니다.
- 정수가 아닌 것이 입력된다면 '잘못된 값을 입력했습니다.'라고 출력하고 게임을 계속 이어나갑니다.

```python
import random
number = random.randint(1, 100)
print(number)
```

```python
while True: 
    try:
        a = int(input(' 1~100까지의 정수 중 하나를 입력하세요.'))
    except Exception as e:
        print('1부터 100까지의 정수 중 하나를 제대로 입력하세요.')
    else:
        if a == number:
            print('정답입니다.')
            break
        else:
            print('잘못된 값을 입력했습니다.')

```

```python
## 선생님 풀이

def guess_game() :
    number = random.randint(1, 100)

    while True:
        try:
            guess = int(input(' 1~100까지의 정수 중 하나를 입력하세요.'))

            if number == guess :
                print('정답입니다.')
                break

            elif guess < 1 or guess > 100:
                print('1부터 100까지의 정수 중 하나를 제대로 입력하세요.')

            elif number > guess:
                print('틀렸습니다. 더 큰 값입니다.')
        
            else : 
                print('틀렸습니다. 더 작은 값입니다.')

        except ValueError:
            print('1부터 100까지의 정수 중 하나를 제대로 입력하세요.')

print(guess_game())
```
