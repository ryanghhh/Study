## **[2] 타입 변환**
`데이터명[컬럼명].astype(타입)`

```python
df['PassengerId']
```

```python
df['PassengerId'].astype(str)  ## 문자열로 출력만함. 본질은 변하지 않았음
```

```python
df1 = df.copy()
df1['PassengerId'] = df1['PassengerId'].astype(str)  ## df1의 PassengerId 열을 문자열로 변환하여 df1에 저장하여 덮어씌우기
df1.info()
```

```python
df['Age']
```

```python
df['Age'].astype(int)  ## 결측치를 정수로 반환할수 없어서 에러가 발생함
```

```python
df['Age'].fillna(-1).astype(int)  ## 결측치를 -1로 바꿔주고 int로 바꾸면 정수로 변환이 됨
```

```python
df1 = df.copy()
df1['Age'] = df1['Age'].fillna(-1).astype(int) ## 실제로 데이터를 바꾸고 싶다면 다시 데이터를 저장하는 이 과정을 거쳐야함.
df1.tail()
```

```python
import numpy as np  ## null 값을 가져오려면 numpy를 import 해줘야함.

df1 = df.copy()
df1['Age'] = df1['Age'].fillna(-1).astype(int).replace(-1, np.nan)  ## age를 정수로 바꾼 다음 -1이 아니라 다시 결측치로 놔두고 싶을 경우, -1을 null 값으로 바꾸는 함수를 하나 더 적어준다. 
df1.tail()
```
