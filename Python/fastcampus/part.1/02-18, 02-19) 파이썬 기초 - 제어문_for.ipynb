## **[1] 파이썬에서 `for`란?**
- 가장 많이 쓰이는 반복문
- 리스트/튜플/문자열의 첫 번째 요소부터 마지막 요소까지 차례로 변수를 받아 반복적으로 수행문을 실행합니다.


    for 변수 in 리스트/튜플/문자열:
        수행문1
        수행문2
        ...

```python
ls = [1,2,3,4,5]
for i in ls:
    print(i)
```

- 위 코드는 아래와 같다.

```python
print(ls[0])
print(ls[1])
print(ls[2])
print(ls[3])
print(ls[4])
```

- 앞서 배웠던 `range` 함수가 `for`문과 함께 많이 쓰인다.

```python
for i in range(1,6):
    print(i)
```

```python
ls = [1,2,3,4,5]
for i in range(len(ls)):
    print(ls[i])
```

```python
list(range(len(ls)))
```

```python
a = ['a','b','c','d']
for i in range(len(a)):
    print(a[i])
```

```python
for i in a:
    print(i)
```

## **[2] `continue, break, pass`**
- `continue`는 해당 조건에서 그 루프가 끝나고 다음 루프로 넘어가도록 합니다.
- `pass`는 실행할 코드가 없는 것과 마찬가지로 해당 루프는 모두 실행하고 다음 루프로 넘어갑니다.
- `break`는 조건문 자체를 중단시킵니다.



```python
for i in range(1, 11):
    if i % 2 == 0:
        continue
    print(i)
```

```python
for i in range(1, 11):
    if i % 2 == 0:
        break
    print(i)
```

```python
for i in range(1, 11):
    if i % 2 == 0:
        pass
    print(i)
```

```python
a = 0
while a < 10:
    a += 1
    if a % 2 == 0:
        break
    print(a)
```

## **[3] 리스트에서 `for`문 쓰기**
- `append` 함수가 유용하다.

- 1~50까지의 5의 배수를 리스트에 담으려면?

```python
ls = []
for i in range(1, 51):
    if i % 5 == 0:
        ls.append(i)

print(ls)
```

## **[5] 튜플에서 `for`문 쓰기**
- 다양하게 인덱싱 할 수 있다.

```python
a = [(1,2), (3,4), (5,6)]
for i in a:
    print(i[0]+i[1])
```

```python
a = [(1,2), (3,4), (5,6)]
for i, j in a:
    print(i + j)
```

```python
a = [[1,2], [3,4], [5,6]]
for i, j in a:
    print(i + j)
```

## **[6] `for`문 중첩**
- 구구단을 작성해보자.

```python
for i in range(2, 10):
    for j in range(1, 10):
        print(i * j, end = ' ')
    print()
```

### 2. 1부터 1000까지 5의 배수의 합을 `for`문으로 구해보세요.

```python
total = 0

for i in range(1,1001) :
    if i % 5 == 0 :
        total += i

print(total)
```

```python
## 선생님 풀이
sum = 0
for i in range(1, 1001) :
    i += 1
    if i % 5 != 0 :
        continue
    sum += i
print(sum)
```

### 3. 아래 리스트는 반 학생들의 수학 점수 리스트입니다. 각 등급별 학생수를 구하세요.
    81~100: A
    61~80: B
    41~60: C
    0~40: D

```python
score_list = [20,100,45,50,60,85,70,75,80,90,95,25,90]
```

```python
dic = {'A':'81=<score_list=<100',
       'B':'61=<score_list=<80',
       'C':'41=<score_list=<60',
       'D':'21=<score_list=<40',}

dic.items()

for k, v in dic.items():
    print(k, ':', v)
```

```python
## GPT 풀이
score_list = [20, 100, 45, 50, 60, 85, 70, 75, 80, 90, 95, 25, 90]

grade_count = {'A': 0, 'B': 0, 'C': 0, 'D': 0}

for score in score_list:
    if 81 <= score <= 100:
        grade_count['A'] += 1
    elif 61 <= score <= 80:
        grade_count['B'] += 1
    elif 41 <= score <= 60:
        grade_count['C'] += 1
    else:
        grade_count['D'] += 1

for grade, count in grade_count.items():
    print(f"{grade} : {count}")
```

```python
## 선생님 풀이

grade_dic = {'A':[], 'B':[], 'C':[], 'D':[]}
for i in score_list:
    if i >= 81 and i <= 100:
        grade_dic['A'].append(i)
    elif i >= 61 and i <= 80:
        grade_dic['B'].append(i)
    elif i >= 41 and i <= 60:
        grade_dic['C'].append(i)
    else:
        grade_dic['D'].append(i)

grade_dic
```

```python
for k, v in grade_dic.items():
    print(f'등급 {k} : {len(v)}명')
```

### 4. 숫자 맞추기 게임을 만들어보세요.
- 맞출 숫자는 아래 코드와 같이 1~100까지 자연수 중 랜덤으로 정합니다.
- 숫자를 입력할 때마다 더 큰지 작은지 여부를 말해줍니다.
- 단, 기회는 5번이고 5번이 지나면 `'GAME OVER, 정답은: 00'`를 출력합니다.


```python
import random
number = random.randint(1, 100)
```

```python
for i in range(1, 6) :
    game = int(input('숫자를 입력하세요.'))
    if game > number : 
        print('입력한 숫자보다 작습니다.')
    elif game < number :
        print('입력한 숫자보다 큽니다.')
    else : 
        print('정답입니다.')
        break
    if i == 5 : 
        print(f'GAME OVER, 정답은 : {number}')
    
```

```python
## 선생님 풀이

for i in range(5) :
    game = int(input('숫자를 입력하세요.'))
    if game > number : 
        print('입력한 숫자보다 작습니다.')
    elif game < number :
        print('입력한 숫자보다 큽니다.')
    else : 
        print('정답입니다.')
        break

if game != number :
    print(f'GAME OVER, 정답은 : {number}')
```
