## **[1] 행, 열 조회**

### **기본적인 방법**
- 행
 - 한 개
    - `데이터프레임명[인덱스:인덱스+1]`
 - 여러 개
    - `데이터프레임명[시작인덱스:끝인덱스+1]`

```python
df.head()
```

```python
df[:10]
```

```python
df[3:7]
```

```python
df[3:4]
```

- 열
 - 한 개
    - `데이터프레임[컬럼명]`
    - `데이터프레임.컬럼명`
 - 여러 개
    - `데이터프레임[[컬럼명1, 컬럼명2, ...]]`

```python
df['Survived']
```

```python
df.Survived
```

```python
df['Survived'].to_frame()
```

```python
df[['Survived']]
```

```python
df[['Survived','Pclass','Name']]  ## 전체 데이터를 불러오고 나서 가공하면서 원하는 열만 가져오고 싶을때는 이런식으로 인덱싱하면 됨 
```

### **loc, iloc 사용하기**

- loc: 레이블값을 사용하여 조회
 - `데이터프레임명.loc[행조건,열조건]`
 - 열만 조회할때는 행조건에 `:`를 입력합니다.

```python
df.loc[3,]
```

```python
df.loc[:,'Name']
```

```python
df.loc[3,'Name']  ## 인덱스가 3인 행에서 Name열의 값을 가져옴
```

```python
df.loc[3:5,]  ## 3~5행을 가져옴
```

```python
df.loc[3:5,['Age','Pclass']]  ## 3~5행에서 Age와 Pclass열을 가져옴
```

```python
df.loc[[1,3,5],['Age','Pclass']] ## 특정 인덱스를 가져오고 싶을때는 리스트로 묶어서 인덱싱하면 됨
```

```python
df1 = df.set_index('Name') #Name열을 인덱스로 설정
df1.head()
```

```python
df1.loc['Heikkinen, Miss. Laina',]
```

```python
df1.loc[['Heikkinen, Miss. Laina','Allen, Mr. William Henry',]]
```

```python
df1.loc[3:5,]  ## 에러뜨는 이유 : lock함수가 위치 인덱스를 가지고 오는 것이 아니라 인덱스 이름을 가지고 오기 때문. 위치로 찾을떄는 iloc을 사용해야함
```

- iloc: 위치인덱스를 사용하여 조회
 - `데이터프레임명.iloc[행인덱스조건,열인덱스조건]`

```python
df1.iloc[3,]
```

```python
df.iloc[3:7,]
```

```python
df.iloc[[1,3,5],] ## 리스트로 인덱싱 가능
```

```python
df1.iloc[3:5,]
```

```python
df.iloc[3:7,2:4]
```

## **[2] 원하는 조건으로 데이터 추출하기**

### **데이터 정렬하기**
- `데이터프레임명.sort_values(정렬기준컬럼)`
- 내림차순으로 정렬하고 싶다면 `ascending=False` 조건을 걸어주면 됩니다.

```python
df.head()
```

```python
df.sort_values('Age') ## 나이순으로 정렬
```

```python
df.sort_values('Age', ascending=False) ## 나이순으로 내림차순 정렬
```

```python
df.sort_values(['Age','Fare'], ascending=[False,True]) ## 나이순으로 내림차순, 나이가 같을 경우 요금순으로 오름차순 정렬
```

### **특정 조건을 충족하는 데이터 추출하기**
- `데이터프레임명[조건식]`
- `데이터프레임명.query('조건식')`



```python
df[df['Pclass'] == 1]
```

```python
df[(df['Pclass'] == 1) & (df['Age'] >= 30)] ## Pclass가 1이고 나이가 30세 이상인 사람들만 가져옴
```

```python
df[(df['Pclass'] == 1) | (df['Age'] >= 30)] ## Pclass가 1이거나 나이가 30세 이상인 사람들만 가져옴
```

```python
df.query('Pclass == 1') ## 꼭 조건 문자열안으로 넣기
```

```python
df.query('Pclass == 1 and Age >= 30')
```

```python
df.query('Pclass == 1 or Age >= 30')
```

```python
df[df['PassengerId'].isin([3,100,500])] ## PassengerId가 3,100,500인 사람들만 가져옴
```

```python
df.query('PassengerId in [3,100,500]')
```

```python
df.query('PassengerId.isin([3,100,500])')
```

```python
import random

passengerid_list = list(df['PassengerId'])
passengerid_sample = random.sample(passengerid_list, 10)
```

```python
df[df['PassengerId'].isin(passengerid_sample)]
```

```python
df.query('PassengerId in @passengerid_sample')
```
