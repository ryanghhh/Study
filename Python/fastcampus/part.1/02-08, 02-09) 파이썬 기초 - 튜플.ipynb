## **[1] 튜플이란?**
- 리스트와 비슷하게 여러 요소들을 묶을 수 있습니다.
- 다만 튜플의 요소는 수정이나 삭제가 불가능합니다.
- 프로그램 실행 중 절대 변하지 않거나 변해서는 안되는 값들이 있다면 튜플로 저장하는 것이 좋습니다.
- 요소들을 괄호(`()`)로 감싸주면 튜플이 됩니다.

```python
a = (1,2,3,4,5)
b = ('life','is','too','short')
c = (1,2,'life','is')
d = (1,2,['life','is',('too','short')])

print(a, type(a))
print(b, type(b))
print(c, type(c))
print(d, type(d))
```

- 요소가 하나일 경우는 ,를 뒤에 붙여주어야 합니다.

```python
x = (3) ## 튜플로 인식 X
```

```python
type(x) ## 튜플로 인식 X
```

```python
x = (3,)
```

```python
type(x)
```

## **[5] 튜플 함수**
- `len`: 튜플의 길이를 구합니다.
- `sum, min, max`: 튜플의 요소들의 합, 최솟값, 최댓값을 구합니다.
- `in`: 특정 요소가 튜플 안에 있는지 확인합니다.
- `count`: 튜플의 특정 요소의 개수를 반환합니다.
- `index`: 튜플에 요소가 있는 경우 인덱스 값을 반환합니다.
- `join`: 튜플의 요소들을 하나의 문자열로 합칩니다.

```python
a = (1,2,2,4,5)
print(len(a))
print(sum(a))
print(min(a))
print(max(a))
print(1 in a)
print(a.count(2))
print(a.index(2))
```

```python
ls = ('a','b','c','d','e')
' '.join(ls)
```

## **[6] 튜플 함수 심화**
- `max(튜플, key = lambda x:x[인덱스])` : 다중 튜플의 경우 정렬 기준을 정의하여 튜플 내 최댓값을 탐색함 ##lambda는 뒤에 오는 것들에 대한 정의 함수
- `min(튜플, key = lambda x:x[인덱스])` : 다중 튜플의 경우 정렬 기준을 정의하여 튜플 내 최솟값을 탐색함

```python
a = (('a', 1), ('b',7), ('c', 3), ('d', -5), ('e', -9))
print(max(a, key = lambda x:x[0])) ## 튜플들 안의 첫번째 자리 기준으로로 max 구하기기
print(max(a, key = lambda x:x[1])) ## 튜플들 안의 두두번째 기준으로로 max 구하기기

print(min(a, key = lambda x:x[0]))
print(min(a, key = lambda x:x[1]))
```

- 두번째 요소의 절댓값을 기준으로 최댓값/최솟값을 탐색하고 싶다면?

```python
print(max(a, key = lambda x:abs(x[1])))  ## 절댓값 함수 abs()
print(min(a, key = lambda x:abs(x[1])))
```

- 리스트에도 비슷하게 적용할 수 있습니다.

```python
b = list(a)
b
```

```python
print(sorted(b, key = lambda x:x[1]))  ## 숫자기준 정렬렬
print(sorted(b, key = lambda x:x[1], reverse=True)) ## 숫자기준 내림차순
print(sorted(b, key = lambda x:abs(x[1])))  ## 절댓값 작은 순서대로로
```
