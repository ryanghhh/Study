## **[1] List Comprehenshion이란?**
- 리스트 안에 `for`문을 포함하여 한 줄로 편하게 코드를 작성할 수 있게 하는 문법입니다.


    [표현식 for 항목 in 반복가능객체 if 조건문]

```python
ls = []
for i in range(1, 5):
    ls.append(i)
print(ls)
```

```python
ls = [i for i in range(1,5)]
print(ls)
```

```python
ls = []
for i in range(1, 5):
    if i % 2 == 0:
        ls.append(i)
print(ls)
```

```python
ls = [i for i in range(1,5) if i % 2 == 0]
print(ls)
```

```python
ls = []
for i in range(1, 5):
    if i % 2 == 0:
        ls.append(i * 4)
print(ls)
```

```python
ls = [i * 4 for i in range(1,5) if i % 2 == 0]
print(ls)
```

- `for`문을 여러 개 사용하는 것도 가능하다.


    [표현식 for 항목1 in 반복가능객체1 if 조건문1
        for 항목2 in 반복가능객체2 if 조건문2
        ...
    ]

```python
gugudan = []
for i in range(2, 10):
    for j in range(1, 10):
        gugudan.append(i*j)
print(gugudan)
```

```python
gugudan = [i*j for i in range(2,10) for j in range(1,10)]
print(gugudan)
```

## **[2] Dict comprehension이란?**
- `list comprehension`과 마찬가지로 딕셔너리도 for문을 간편하게 이용할 수 있다.

```python
names = ['Merry','John','Chris']
num = [1,2,3]

dic = {names[i]:num[i] for i in range(len(names))}
print(dic)
```

```python
dic1 = {}
for i in range(len(names)):
    dic1[names[i]] = num[i]

dic1
```

```python
names = ['Merry','John','Chris']
num = [1,2,3]

dic = {k:v for k, v in zip(names, num)}
print(dic)
```

```python
dic1 = {}
for k, v in list(zip(names, num)):
    dic1[k] = v

dic1
```

```python
names = ['Merry','John','Chris']
num = [1,2,3]

dic = {k:v*5 for k, v in zip(names, num)}
print(dic)
```

```python
dic1 = {}
for k, v in list(zip(names, num)):
    dic1[k] = v*5

dic1
```

```python
names = ['Merry','John','Chris']
num = [1,2,3]

dic = {k:v*5 for k, v in zip(names, num) if v % 2 == 1}
print(dic)
```

```python
dic1 = {}
for k, v in list(zip(names, num)):
    if v % 2 == 1:
        dic1[k] = v*5

dic1
```

### 3. 다음 리스트는 리스트 안에 리스트가 들어 있습니다. 리스트 안의 리스트를 하나의 리스트로 만들어보세요.
    [[1,2],[3,4],[5,6]]
    ->
    [1,2,3,4,5,6]

```python
ohlc = [[100, 110, 70, 100],
        [200, 210, 180, 190],
        [300, 310, 300, 310]]
```

```python
flat_list = [x for row in ohlc for x in row]
print(flat_list)
```

```python
flat_list = []
for row in ohlc:
    for x in row:
        flat_list.append(x)
flat_list
```

```python
## 선생님 풀이
ls = []
for i in ohlc:
    for j in i :
        ls.append(j)
print(ls)
```

```python
[ j for i in ohlc for j in i ]
```

### 4. 다음 리스트의 요소의 빈도값을 `dict comprehension`을 사용하여 딕셔너리로 출력해보세요.

```python
x = ['a','a','b','b','b','c','c','c','c']
```

```python
freq = {i: x.count(i) for i in sorted(set(x))}
print(freq)
```

```python
## 선생님 풀이

x.count('a')
x.count('b')
x.count('c')
```

```python
dix = {}
for i in set(x) : 
    dix[i] = x.count(i)
dix
```
