````python
import folium
import base64

# 이미지 경로
image_path_1 = r"C:\Users\ryang\OneDrive\사진\카메라 앨범\WIN_20250409_22_38_04_Pro.jpg"
image_path_2 = r"C:\Users\ryang\OneDrive\사진\카메라 앨범\WIN_20250503_18_20_14_Pro.jpg"

# 이미지1 base64 인코딩
with open(image_path_1, 'rb') as f1:
    encoded1 = base64.b64encode(f1.read()).decode()

# 이미지2 base64 인코딩
with open(image_path_2, 'rb') as f2:
    encoded2 = base64.b64encode(f2.read()).decode()

# HTML 팝업
html1 = f'<img src="data:image/jpeg;base64,{encoded1}" width="200">'
html2 = f'<img src="data:image/jpeg;base64,{encoded2}" width="200">'

# 지도 생성
m = folium.Map(location=[35.13, 129.11], zoom_start=15)

# 마커 1
folium.Marker(
    [35.129568473244156, 129.1103649881546],
    tooltip='우리집',
    popup=folium.Popup(html1, max_width=300),
    icon=folium.Icon(color='red', icon='star')
).add_to(m)

# 마커 2
folium.Marker(
    [35.1819163854354, 129.1273588453472],
    tooltip='쫑아님집',
    popup=folium.Popup(html2, max_width=300),
    icon=folium.Icon(color='blue', icon='star')
).add_to(m)

# 지도 출력
m

````

````python
folium.CircleMarker([35.129568473244156, 129.1103649881546]
              , color = 'red'
              , radius = 50).add_to(m)
m
````
