## **[1] 파이썬에서 `if`란**
조건의 참과 거짓을 판단해 그에 따라 동작을 수행합니다.

    날씨가 맑으면 공원에 가고, 날씨가 흐리면 카페에 가고, 그 외의 날씨에는 집에 있는다.

이것을 파이썬으로 표현하려면? `if, elif, else` 조건문을 사용합니다.

    if 날씨가 맑다:
        공원에 간다
    elif 날씨가 흐리다:
        카페에 간다
    else:
        집에 있는다


`if` 문의 기본 구조는

    if 조건문:
        수행문1 ## 들여쓰기
        수행문2
        ...

이고, 조건이 여러개라면 `elif`와 `else`를 사용합니다.

    if 조건문1:
        수행문1
        수행문2
        ...
    elif 조건문2:
        수행문1
        수행문2
        ...
    ...
    else:
        수행문1
        수행문2
        ...


```python
x = 3
if x == 3:
    print('3입니다.')
else:
    print('3이 아닙니다.')
```

```python
x = -1
if x >= 0:
    print('양수')
elif x < 0:
    print('음수')
else:
    print('숫자가 아닙니다.')
```

- `elif`는 아래와 같이 `else`문 안에서 조건문이 복잡해지는 것을 단순화한 것입니다.

```python
x = 3
if x >= 0:
    print('양수')
else:
    if x < 0:
        print('음수')
    else:
        print('숫자가 아닙니다.')
```

## **[3] 연산자**
|연산자|설명|
|:---------|:---|
|A > B|A가 B보다 크다|
|A < B|A가 B보다 작다|
|A >= B|A가 B보다 크거나 같다|
|A <= B|A가 B보다 작거나 같다|
|A == B|A와 B가 같다|
|A != B|A가 B와 같지 않다|
|A and B|A와 B 둘 다 참이어야 참이다|
|A or B|A와 B 중 하나만 참이어도 참이다|
|not A|A가 거짓일 때 참이다|
|x in 리스트/튜플/문자열|x가 리스트/튜플/문자열에 들어있을 때 참이다|
|x not in 리스트/튜플/문자열|x가 리스트/튜플/문자열에 들어있지 않을 때 참이다|

```python
a = 3
b = 1

print((a == b) and (a >= b))
print((a >= b) and (a <= b))
print((a <= b) and (a != b))
print((a == b) or (a >= b))
print((a >= b) or (a <= b))
print((a <= b) or (a != b))
print(not a == b)
```

```python
a = [1,2,3]
print(1 in a)
print(4 in a)
print(1 not in a)
print(4 not in a)
```

```python
s = 'Hello, World!'
print('h' in s)
print('!' in s)
print('W' not in s)
```

### 4. 아래의 리스트는 어떤 반의 수학 점수 모음입니다. 학생 점수를 입력 받으면 아래와 같이 출력되게끔 코드를 작성하세요.

    '학생의 수학 점수는 00점입니다.'
    '반 평균은 00점이고, 평균보다 00점 낮습니다/00점 높습니다/평균과 같습니다.'

- 만약 학생 점수로 0~100 사이의 정수가 입력되지 않으면 아래와 같이 출력하도록 하세요.


    '잘못된 값을 입력했습니다.'

- 반 평균은 소숫점 둘째 자리에서 반올림하세요.

- 참고 함수
 - 반올림: `round(값, 반올림할 자리수)`
 - 절댓값: `abs(값)`



```python
score_list = [70,90,50,30,100,95,95,40,90,20,100]
```

```python
a = int(len(score_list))
b = int(sum(score_list))
print(a)
print(b)
c = b/a
```

```python
average_ab = round(c, 2)
print(average_ab)
type(average_ab)
```

```python
score = input ('0~100점 사이의 점수를 입력하세요.')
print(f'학생의 수학 점수는 {score}점 입니다.')
my_score = int(score)
if my_score > average_ab :
    print(f'반 평균은 {average_ab}점이며, 평균보다 {round(my_score - average_ab, 2)}점 높습니다.')
elif my_score < average_ab :
    print(f'반 평균은 {average_ab}점이며, 평균보다 {round(average_ab - my_score, 2)}점 낮습니다.')
else :
    print(f'반 평균은 {average_ab}점이며 평균과 같습니다.')
```

```python
## 선생님 풀이
x = int(input("0~100 사이의 정수를 입력해주세요."))
```

```python
score_mean = sum(score_list) / len(score_list)
score_mean = round(score_mean, 2)
score_mean
```

```python
diff = round(abs(x - score_mean))  ## 점수차이를 절댓값으로 변환
```

```python
print(f'학생의 수학 점수는 {x}점 입니다.')
print(f'반 평균은 {score_mean}점이고, ', end = '') ## end문 쓰면 다음 것도 한줄로 쓸수 있음

if x < 0 or x > 100 :
    print('잘못된 값을 입력했습니다.')
elif x > score_mean : 
    print(f'평균보다 {diff}점 높습니다.')
elif x < score_mean : 
    print(f'평균보다 {diff}점 낮습니다.')
else :
    print('평균과 같습니다.')
```

### 5. 가위바위보 게임을 만들어보세요.
- 결과 예시


    '플레이어는 가위를 냈습니다.'
    '컴퓨터는 보를 냈습니다.'
    '플레이어가 승리'

- 플레이어의 값은 `input` 함수로 입력을 받습니다.
- 컴퓨터의 값은 아래의 코드를 이용합니다.


```python
import random
computer = random.choice(['가위', '바위', '보'])

game = input('가위,바위,보를 입력하세요')
if game != '가위' and game !='바위' and game !='보' :
     print('바르게 입력해주세요.')
else :
    print(f'플레이어는 {game}를 냈습니다.')
    print(f'컴퓨터는 {computer}를 냈습니다.')

    if game == computer :
        print('비겼습니다.')
    elif game == '가위' and computer == '바위' :
        print('졌습니다.')
    elif game == '가위' and computer == '보' :
        print('플레이어가 승리')
    elif game == '바위' and computer == '보' :
        print('졌습니다.')
    elif game == '바위' and computer == '가위' :
        print('플레이어가 승리')
    elif game == '보' and computer == '가위' :
        print('졌습니다.')
    elif game == '보' and computer == '바위' :
        print('플레이어가 승리') 
```

```python
## 선생님 풀이
player = input('가위, 바위, 보 중 하나를 입력해주세요.')
```

```python
print(f'플레이어는 {player}를 냈습니다.')
print(f'컴퓨터는 {computer}를 냈습니다.')
```

```python
if player == computer : 
    print('무승부')
elif player == '가위' :
    if computer == '바위' :
        print('컴퓨터 승리')
    else:
        print('플레이어 승리')
elif player == '바위' :
    if computer == '보' :
        print('컴퓨터 승리')
    else:
        print('플레이어 승리')
elif player == '보' :
    if computer == '가위' :
        print('컴퓨터 승리')
    else:
        print('플레이어 승리')
else :
    print('잘못 입력했습니다.')
```

```python
## 딕셔너리 활용 다른 풀이
win_dict = {'가위' : '보', '바위' : '가위', '보':'바위' }

player = input('가위, 바위, 보 중 하나를 입력해주세요.')
print(f'플레이어는 {player}를 냈습니다.')
print(f'컴퓨터는 {computer}를 냈습니다.')

if player not in win_dict :
    print('잘못 입력했습니다.')
elif computer == player :
    print('무승부')
elif win_dict[computer] == player :
    print('컴퓨터 승리')
else :
    print('플레이어 승0리')
```
