## **[2] `while`문 강제로 빠져나가기**

- `while`문은 조건이 거짓이 되면 끝나게 되는데, 중간에 특정 조건에 다다르면 빠져나가고 싶을 때는 `break`를 쓰면 됩니다.

```python
jajang = 0
while True: ## 언제나 참
    jajang += 1
    print(f'짜장면을 {jajang}그릇 먹었습니다.')

    if jajang == 10:
        print('무료 쿠폰을 받았습니다.')
        break ## 10이되면 종료되게
```

- 위의 경우에 `break`문이 없으면 무한루프가 됩니다.

```python
jajang = 0
while True:
    jajang += 1
    print(f'짜장면을 {jajang}그릇 먹었습니다.')

    if jajang == 10:
        print('무료 쿠폰을 받았습니다.')
```

## **[3] 특정 조건에서 수행문을 건너뛰고 계속 `while`문을 반복하려면?**
- 1~10까지의 숫자 중 홀수만 출력하는 코드는 다음과 같이 짤 수 있습니다.


- `continue`는 해당 조건에서 그 루프가 끝나고 다음 루프로 넘어가도록 합니다.

```python
a = 0
while a < 10:
    a += 1
    if a % 2 == 0:
        continue ## 짝수이면 건너뛰어라
    print(a)
```

- 전 시간에 배운 `pass`와의 차이점을 주의합니다. `pass`는 실행할 코드가 없는 것과 마찬가지로 해당 루프는 모두 실행하고 다음 루프로 넘어갑니다.
pass는 해당 조건만 pass
continue는 해당 조건에 다다르면 그 루프를 중단하고 다음 루프로 넘어감

```python
a = 0
while a < 10:
    a += 1
    if a % 2 == 0:
        pass
    print(a)
```

```python
a = 0
while a < 10:
    a += 1
    if a % 2 == 0:
        pass
    else:
        print(a)
```

- `break`는 while문 자체를 중단시킵니다.

```python
a = 0
while a < 10:
    a += 1
    if a % 2 == 0:
        break ## 2 되니까 바로 멈춰버림춰버림
    print(a)
```

### 2. 1부터 1000까지 5의 배수의 합을 구해보세요.

```python
c = 1
total_c = 0

while c <= 1000:
    if c % 5 == 0:      # 5의 배수인지 확인
        total_c += c    ## total_c라는 변수에 c를 더해서 다시 total_c에 저장한다는 뜻
    c += 1              # 꼭 증가시켜줘야 함!

print(total_c)
```

```python
coffee_num = 5
coffee_price = 300

money = input('돈을 넣어주세요.')

if money == 300 :
    print('커피를 줍니다.')
    print(f'남은 커피의 양은 5-{}잔 입니다.')

if money < 300:
    print('무료 쿠폰을 받았습니다.')
```

```python
## 지피티 풀이
coffee = 5  # 커피 재고
price = 300  # 커피 한 잔 가격

while coffee > 0:
    money = int(input("돈을 넣어주세요: "))

    if money == price:
        print("커피를 줍니다.")
        coffee -= 1
        print(f"남은 커피의 양은 {coffee}잔입니다.")
    elif money > price:
        change = money - price
        print(f"거스름돈 {change}원을 주고 커피를 줍니다.")
        coffee -= 1
        print(f"남은 커피의 양은 {coffee}잔입니다.")
    else:
        print(f"커피의 가격은 {price}원입니다. 돈을 다시 돌려주고 커피를 주지 않습니다.")
        print(f"남은 커피의 양은 {coffee}잔입니다.")

print("커피가 다 떨어졌습니다. 판매를 중지합니다.")
```

### 4. while문을 이용하여 아래와 같이 출력해보세요.

```python
i = 8
while i > 0:
    i -= 1

    if i % 2 == 0:
        continue

    star = i
    blank = int((7-i)/2)

    print(' '*blank, '*'*i, ' '*blank, sep=' ')
```
