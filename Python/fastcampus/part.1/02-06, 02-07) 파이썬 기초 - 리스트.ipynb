## **[1] 리스트란?**
- 요소들의 모음을 나타내는 자료형
- 요소들을 대괄호(`[]`)로 감싸주어 만들 수 있습니다.
- 모든 자료형을 리스트의 요소로 담을 수 있습니다.

```python
a = [1,2,3,4,5]
b = ['life','is','too','short']
c = [1,2,'life','is']
d = [1,2,['life','is',['too','short']]]

print(a, type(a))
print(b, type(b))
print(c, type(c))
print(d, type(d))
```

```python
type([1])
```

## **[3] 리스트 연산**
- 더하기와 곱하기를 할 수 있습니다.
 - `+`: 리스트와 리스트를 순서대로 연결하여 하나의 리스트를 만듭니다.
 - `*`: 리스트와 숫자형을 곱하여 리스트 안의 요소를 숫자만큼 반복한 하나의 리스트를 만듭니다.

```python
a = [1,2,3]
b = [4,5,6]
a + b
```

```python
a * 3
```

- 리스트에 다른 자료형을 더하면 오류가 납니다.

```python
a + 4
```

```python
a + [4]
```

## **[4] 리스트 값 치환**
- 리스트는 인덱싱을 통해 값을 수정할 수 있습니다.

```python
a = [1,2,3,4,5]
print(a)
a[0] = 0
print(a)
```

```python
a = [1, [2, 3], [4, 5, [6, 7]]]

a[0] = 10
print(a)

a[1][0] = [9, 10, 11]
print(a)

a[2][-1] = 8
print(a)

a[1][0][1:] = [20, 30]
print(a)
```

```python
print(a[1][0][1:])
a[1][0][1:] = 50 ## 슬라이스에 값을 넣을떄는 리스트로 넣어줘야 한다. [50] 처럼.
```

## **[5] 리스트 함수**
- `len` 함수: 리스트의 길이를 구합니다.
- `sum, min, max` 함수: 리스트 요소들의 합, 최솟값, 최댓값을 구합니다.
- `in`: 특정 요소가 리스트 안에 있는지 확인합니다.
- `append, extend`: `append`는 리스트의 맨 마지막에 요소를 추가, `extend`는 리스트의 맨 마지막에 다른 리스트를 추가합니다.
- `insert`: 리스트의 특정 위치에 특정 요소를 삽입합니다.
- `remove, pop`: 리스트의 요소를 삭제합니다.
- `count`: 리스트의 특정 요소의 개수를 반환합니다.
- `index`: 리스트에 요소가 있는 경우 인덱스 값을 반환합니다.
- `sort, sorted`: 리스트를 정렬합니다.
- `join`: 리스트의 요소들을 하나의 문자열로 합칩니다.
- `split`: 문자열을 리스트의 요소들로 쪼갭니다.

### `len`

```python
a = [1,2,3,4,5]
len(a)
```

### `sum, min, max`

- 리스트가 숫자형으로 이루어졌을 때 요소들의 평균을 구하려면?

```python
sum(a)/len(a)
```

- 최솟값과 최댓값도 간단하게 구할 수 있습니다.

```python
print(min(a))
print(max(a))
```

- 리스트의 요소가 숫자가 아니라면?

```python
a = ['a','b','c','d','e']
print(sum(a))
```

```python
print(min(a))
print(max(a))
```

### 3. 다음 리스트에서 중앙값을 구하세요.
- 중앙값 구하는 방법
 1.   데이터를 작은 수부터 큰 수 순서대로 정렬
 2.   데이터의 개수가 홀수라면 정렬된 결과의 가운데 수
 3.   데이터의 개수가 짝수라면 정렬된 결과의 가운데 두 수의 평균



```python
a = [1,100,90,32,78,40,2]
```

```python
len(a)
a.sort()
a[3]
```

```python
##선생님 풀이
a.sort()
len(a)%2==1 ##홀수라고 판명
center_index = len(a)//2 ## a의 길이를 2로 나눈 몫을 가지고 오면 가운데 수의 인덱스를 구할 수 있다 -> 3
center_index ## ->3
a[center_index] ## -> a[3] -> 40
```
