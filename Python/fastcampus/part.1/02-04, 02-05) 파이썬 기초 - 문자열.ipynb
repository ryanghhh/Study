## **[3] 인덱싱**
- 인덱스는 `[]`를 통해 지정할 수 있으며, 첫 번째 위치는 `0`이고, 맨 마지막 위치는 `-1`로도 표현할 수 있습니다.
- `:`를 통해 슬라이싱을 할 수 있습니다.
 - `변수[시작 인덱스:끝 인덱스:간격]`
- 공백도 한 칸을 차지합니다.

```python
a = 'Hello, World'
print(a[0])
print(a[-1])
print(a[5])
```

```python
print(a[:])
print(a[3:7])
print(a[:-1])
print(a[-5:])
```

```python
print(a[::2])
print(a[1:-1:3])
```

## **[5] 문자열과 숫자형의 혼용**
- 문자열과 숫자형을 더하면 에러가 납니다.
- 숫자형을 문자열으로 변형한 후 더해주어야 합니다.
 - str(숫자형 변수)

```python
a = 'Number'
b = 7
print(a, b)
```

```python
print(a + b) ## a는 글자고 7은 숫자기 때문에 안더해짐짐
```

```python
print(a + str(b)) ## 숫자7을 문자로 바꿔준다
```

- 숫자형을 문자열로 변환할 수 있는 것처럼, 문자열도 숫자형으로 변환할 수 있습니다.
 - 정수형: `int`
 - 실수형: `float`

```python
a = "12345"
print(int(a) + 1)
```

```python
a = "12345.123"
print(float(a) + 1)
```

- 변환할 수 없는 경우에는 에러가 납니다.

```python
a = 'abcd12'
int(a)
```

```python
a = '1234.5'
int(a)
```

## **[7] 문자열 함수**
- `upper`: 대문자로 변환한 값을 출력
- `lower`: 소문자로 변환한 값을 출력
- `lstrip, rstrip, strip`: 공백 제거한 값을 출력
- `count`: 문자열 안의 특정 문자열 개수 세기
- `find`: 문자열 안의 특정 문자열이 처음 나온 위치 반환
 - 문자열이 없으면 -1을 반환
- `index`: `find`와 같으나 문자열이 없으면 오류 발생
- `join`: 특정 문자열을 각각의 문자 사이에 삽입하여 출력
- `split`: 특정 문자열을 기준으로 문자열을 나눈 값을 출력
- `replace`: 특정 문자열을 특정 문자열로 반환
- `startswith, endswith`: 특정 문자열로 시작하는지/끝나는지 확인

### `upper, lower`

```python
a = 'pie'
a.upper()
```

- 문자열 함수들은 변수의 값을 변경하는 것이 아니라 함수의 결과물을 출력하는 역할을 하므로,
- 변수의 값을 변경하고 싶으면 변수에 값을 덮어씌워 주어야 합니다.

```python
print(a)
```
